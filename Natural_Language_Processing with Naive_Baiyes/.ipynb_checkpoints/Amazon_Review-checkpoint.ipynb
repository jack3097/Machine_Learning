```python
# Importing all the required libraries
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
df = pd.read_csv('amazon_train.csv')
df.head()
```

```python
df.info()
```

```python
wordnet = WordNetLemmatizer()
def pre_process(msg,corpus):
    msg = msg.lower()
    msg = re.sub('[^a-zA-Z]',' ', msg)
    msg = nltk.word_tokenize(msg)
    msg = [wordnet.lemmatize(word) for word in msg if word not in set(stopwords.words('english'))]
    msg = ' '.join(msg)
    corpus.append(msg)
```

```python
corpus1 = []
df['Review Text'].apply(lambda x : pre_process(x,corpus1))
corpus1
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
x1 = tfidf.fit_transform(corpus1).toarray()
x1 = pd.DataFrame(x1)
x1.head()
```

```python
corpus2 = []
df['Review Title'].apply(lambda x : pre_process(x,corpus2))
corpus2
```

```python
tfidf = TfidfVectorizer()
x2 = tfidf.fit_transform(corpus2).toarray()
x2 = pd.DataFrame(x2)
x2.head()
```

```python
train = pd.concat([x1,x2],axis=1)
```

```python
target = df['topic']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
target = le.fit_transform(target)
target = pd.DataFrame(target)
```

```python
target.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(train,target,test_size=.30,random_state=1)
print("x_train shape:",x_train.shape)
print("y_train shape:",y_train.shape)
print("x_test shape:",x_test.shape)
print("y_test shape:",y_test.shape)
```

```python
#training model
from sklearn.naive_bayes import MultinomialNB
mNB = MultinomialNB().fit(x_train,y_train)
```

```python
y_pred = mNB.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,precision_score
print("accuracy score:",accuracy_score(y_test,y_pred))
print("precision score:",precision_score(y_test,y_pred,average='micro'))
```
