```python
# Importing all the required libraries
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
#from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
df = pd.read_csv('out.csv')
```

```python
df1 = df[['Message']]
```

```python
df1.head()
```

```python
wordnet = WordNetLemmatizer()
def pre_process(msg,corpus):
    msg = msg.lower()
    msg = re.sub('[^a-zA-Z]',' ', msg)
    msg = nltk.word_tokenize(msg)
    msg = [wordnet.lemmatize(word) for word in msg if word not in set(stopwords.words('english'))]
    msg = ' '.join(msg)
    corpus.append(msg)
```

```python
corpus1 = []
df1['Message'].apply(lambda x: pre_process(x,corpus1))
corpus1
```
